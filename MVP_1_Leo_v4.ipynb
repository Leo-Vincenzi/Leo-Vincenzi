## 1.0 Definição do Problema
---

**Qual o objetivo desse MVP?**

O desafio desse projeto acadêmico é, com os conhecimentos adquiridos nessa sprint, desenvolver uma solução que busque o aumento das vendas e da lucratividade de uma empresa comercial do ramo varejista de roupas, agasalhos, calçados e acessórios do ramo de motocicletas, a qual possui lojas físicas em todos os estados dos EUA.

**Como fazer isso?**

Por meio de uma idéia e uma base de dados de vendas, obteremos insights e informações que possam ser relevantes para melhorar a tomada de decisão gerencial. Tentaremos antecipar o comportamento do consumidor, buscando produzir um modelo capaz de prever qual o tipo de produto nosso cliente estará mais propenso a comprar da próxima vez em que aparecer numa de nossas lojas.

Inicialmente examinaremos melhor o perfil do nosso cliente, suas preferências, tentando entender melhor porque um estado vende mais do que o outro, sua sazonalidade e quais são os produtos mais vendidos.

Desenvolveremos, então, um modelo capaz de entender o comportamento do consumidor, otimizar a comunicação por meio de campanhas de vendas direcionadas, remover do leque de produtos aqueles de baixa saída, melhorar o gerenciamento de estoques com previsões de vendas mais acertadas, lançar novos produtos mais rapidamente, definir preços e promoções ideais, com descontos personalizados de acordo com a geolocalização da venda, ou recomendando alternativas de pagamento e políticas de preço, considerando os riscos e a rentabilidade. Também poderemos contribuir com a prevenção de fraudes.

A meta é reduzir custos melhorando a performance na carteira de clientes.

**Sobre o dataset utilizado**

O dataset usado neste projeto será o ***Customer Shopping Trends Dataset***. Ele mostra cada item vendido, e oferece informações a respeito do comportamento do consumidor e seus padrões de compra. Como seus dados estão rotulados, trata-se de um **aprendizado supervisionado**.

Este conjunto de dados captura uma gama de atributos do cliente que inclui idade, sexo, histórico de vendas, métodos de pagamento, frequência de vendas e avaliações de *feedback* do cliente.

Além disso, possui dados sobre os tipos de itens adquiridos, épocas preferidas e interações do cliente com ofertas promocionais.

**Premissas sobre o mercado do comércio varejista de vestuário do segmento de motocicletas**

Compreender as dinâmicas e tendências do cliente é fundamental para que a gestão das lojas entenda melhor o seu nicho de mercado, oferecendo novos produtos, desenvolvendo estratégias de marketing eficazes, e alinhando suas estratégias de vendas com as novas preferências dos consumidores.

Por ser um segmento volátil, sazonal e ligado a modismos, a análise de dados pode se tornar um diferencial estratégico vantajoso para empresas ligadas ao comércio varejista em uma tomada de decisão gerencial rápida e precisa.

18. **Frequencia_Vendas** - Frequência em que o cliente retorna à loja.

Utilizaremos agora o coeficiente de correlação de Pearson para a nossa próxima análise. Esse coeficiente é uma medida adimensional, que assume valores no intervalo entre -1 e +1, e mede a intensidade e a direção de relações lineares. A intensidade diz respeito ao grau de relacionamento entre duas variáveis, e a direção indica se essa correlação é positva ou negtiva.



> *OBS: Interpretando o coeficiente de Pearson: valores acima de 0.9, positivos ou negativos, indicam uma correlação muito forte; valores 0.7 a 0.9, positivo ou negativo, indicam uma correlação forte; valores 0.5 a 0.7, positivo ou negativo, indicam uma correlação moderada; valores 0.3 a 0.5, positivo ou negativo, indicam uma correlação fraca; valores 0.0 a 0.3, positivo ou negativo, indicam uma correlação desprezível.*



Usando o coeficiente de correlação de Pearson, verificaremos a seguir qual é a correlação entre a idade de um cliente e total de suas compras anteriores.

```
# Correlação entre a idade de um cliente e o total de Vendas_Anteriores, usando método Pearson
#
correlacao_idade_vendas_anteriores = dataset['Idade'].corr(dataset['Vendas_Anteriores'], method='pearson')
print(f"O coeficiente de correlação de Pearson entre Idade e Vendas_Anteriores é {correlacao_idade_vendas_anteriores:,.3f}")
#print(f"A idade média dos clientes é {media_de_idade_do_cliente:,.0f} anos")
```

Com um coeficiente próximo a ZERO, verificamos que praticamente não existe correlação entre esses dois atributos.

Para verificar se existe alguma correlação de atributos de um dataset, a matriz de correlação de Pearson costuma ser a ferramenta mais apropriada, já que ela mostra o grau de relação linear entre cada par de variáveis, no formato de uma matriz bidimensional. Como vimos antes, os seus valores podem ter variar entre -1 e +1. Se os dois atributos tendem a aumentar e diminuir juntas, o valor de sua correlação é positivo. Se um atributo de coluna aumenta enquanto o outro diminui, o valor de correlação é negativo. Quando está póximo de zero, não há correlação entre os atributos.

```
# Criando uma matriz de Correlação com MatPlotLib Seaborn
#
sns.heatmap(dataset.corr(), annot=True, cmap='RdBu');
```

Ao visualizar as correlações entre os atributos de coluna através da matriz de correlação, confirmamos o valor do coeficiente obtido anterirmente para "Idade e Vendas_Anteriores", e perceberemos que não parece haver qualquer correlação entre nenhum dos outros atributos, já que a maioria dos coeficientes é zero, ou muito próximo disso.

Se houvesse algum campo azul fora da diagonal, a matriz sugeriria que os atributos de coluna poderiam ter alguma proximidade, mostrando quais possuem maior correlação entre si.

Ou seja, ainda existem atributos de coluna com valores nulos, confirmando que são "Valor_Venda" e "Vendas_Anteriores".

Decidimos substituir esses campos não preenchidos de 'Valor_Venda' e 'Vendas_Anteriores' pela mediana dos restantes valores das colunas.

A vantagem da mediana sobre a média é que a mediana pode fornecer um valor típico do conjunto de dados, não sendo afetada por valores extremos.

```
# Recuperando os nomes das colunas
colunas = list(dataset_limpo.columns)

# Calculando a média dos atributos de coluna "Vendas_Anteriores" e "Valor_Venda"
#
media_de_vendas_anteriores = int(dataset_limpo [ 'Vendas_Anteriores' ].median())
media_de_valor_vendas = dataset_limpo [ 'Valor_Venda' ].median()

# Substituindo os NULL pelas médias correspondentes do restante dos atributos
#
dataset_limpo [ 'Vendas_Anteriores' ].fillna(media_de_vendas_anteriores, inplace=True)
dataset_limpo [ 'Valor_Venda' ].fillna(media_de_valor_vendas, inplace=True)

# Exibindo visualização matricial da nulidade do dataset provisório
#
ms.matrix(dataset_limpo)
```

Com esse resultado confirmamos a não existência de atributos com valores nulos ou zerados em nosso dataset.

A seguir retiraremos a 7ª coluna do dataset ("Localizacao"), já que categorizamos ela no atributo de coluna "Regiao".

```
# O dataset provisório passará a conter todas as colunas, com exceção da 7ª ("Localização")
#
dataset_limpo = dataset_limpo.drop( "Localizacao" , axis = 1)
#
dataset_limpo.info()
```

Por causa dos campos *missing*, os valores dos atributos de coluna "ID_Cliente" e de "Vendas_Anteriores" (que são int64) foram interpretados como float64.

Também verificamos que os atributos de coluna "Sexo", "Categoria", "Tamanho", "Forma_Pagamento" e "Frequencia_Vendas" são do tipo "category".

Vamos resolver esses problemas alterando seus formatos (*Dtype*).

```
# Alterando o formato dos atributos das colunas "ID_Cliente" e de "Vendas_Anteriores" para tipo "int64" no dataset novo
#
dataset_limpo [ 'ID_Cliente' ] = dataset_limpo [ 'ID_Cliente' ].astype( 'int64' )
dataset_limpo [ 'Vendas_Anteriores' ] = dataset_limpo [ 'Vendas_Anteriores' ].astype( 'int64' )

# Alterando o formato dos atributos das colunas "Sexo", "Categoria", "Tamanho", "Forma_Pagamento" e "Frequencia_Vendas" para tipo "category" no dataset novo
#
dataset_limpo [ 'Sexo' ] = dataset_limpo [ 'Sexo' ].astype( 'category' )
dataset_limpo [ 'Categoria' ] = dataset_limpo [ 'Categoria' ].astype( 'category' )
dataset_limpo [ 'Tamanho' ] = dataset_limpo [ 'Tamanho' ].astype( 'category' )
dataset_limpo [ 'Frequencia_Vendas' ] = dataset_limpo [ 'Frequencia_Vendas' ].astype( 'category' )

dataset_limpo.info()
```

Dessa forma somente o atributo de coluna "Valor_Venda" do nosso dataset permanecerá com o formato float64.

Iremos agora substituir os valores dos atributos de coluna "Assinatura", "Desconto" e "Codigo_Promocional", que inicialmente possuíam os valores "Não" e "Sim" (object), pelos valores por "False" e "True" (bool), respectivamente.

```
# Alterando os valores de "Não" e "Sim" (object) para False e True (bool)
#
novos_valores = {'Sim' : 'true', 'Nao' : 'false'}
dataset_limpo [ 'Assinatura' ] = dataset_limpo [ 'Assinatura' ].map(novos_valores)
dataset_limpo [ 'Desconto' ] = dataset_limpo [ 'Desconto' ].map(novos_valores)
dataset_limpo [ 'Codigo_Promocional' ] = dataset_limpo [ 'Codigo_Promocional' ].map(novos_valores)

# Alterando o formato dos atributos de coluna "Assinatura", "Desconto" e "Codigo_Promocional" no dataset novo

dataset_limpo [ 'Assinatura' ] = dataset_limpo [ 'Assinatura' ].astype('bool')
dataset_limpo [ 'Desconto' ] = dataset_limpo [ 'Desconto' ].astype('bool')
dataset_limpo [ 'Codigo_Promocional' ] = dataset_limpo [ 'Codigo_Promocional' ].astype('bool')

# Exibindo as suas 10 primeiras linhas
#
dataset_limpo.head(10)
```
